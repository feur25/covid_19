```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

df = pd.read_csv('owid-covid-latest.csv')

selected_countrie = ['France', 'Finland', 'Fiji', 'China', 'Egypt']

def plot_selected_countries(selected_countries):
    selected_data = df[df['location'].isin(selected_countrie)]
    
    selected_data = selected_data.sort_values(by='total_deaths', ascending=False)
    
    plt.figure(figsize=(10, 5))
    
    if len(selected_data) > 15:
        bars = plt.barh(selected_data['location'], selected_data['total_deaths'])
        plt.ylabel('Pays')
        plt.xlabel('Nombre de morts')
        plt.title('Nombre de morts par pays')
        for bar, death_count in zip(bars, selected_data['total_deaths']):
            if pd.notna(death_count):
                plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, str(int(death_count)), va='center')
        plt.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
    else:
        bars = plt.bar(selected_data['location'], selected_data['total_deaths'])
        plt.xlabel('Pays')
        plt.ylabel('Nombre de morts')
        plt.title('Nombre de morts par pays')
        plt.xticks(rotation=45)
        for bar, death_count in zip(bars, selected_data['total_deaths']):
            if pd.notna(death_count):
                plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(int(death_count)), ha='center', va='bottom')
        plt.tick_params(axis='y', which='both', left=False, labelleft=False)
    
    plt.box(on=None)
    
    plt.show()

all_countries = sorted(df['location'].unique())

country_selector = widgets.SelectMultiple(
    options=all_countries,
    value=['France', 'Finland', 'Fiji', 'China', 'Egypt'],  
    description='Pays:',
    disabled=False
)

def update_plot(change):
    new_country = change.new
    if isinstance(new_country, tuple): 
        new_country = new_country[0] 
    selected_countrie.append(new_country)

country_selector.observe(update_plot, 'value')

interact(plot_selected_countries, selected_countries=country_selector);
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['location'], df['new_deaths'])
plt.xlabel('Pays')
plt.ylabel('Nombre de morts')
plt.title('Nombre de nouveaux morts par pays')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 5))
plt.bar(df['location'], df['new_cases'])
plt.xlabel('Pays')
plt.ylabel('Nombre de cas')
plt.title('Nombre de nouveaux cas par pays')
plt.xticks(rotation=45)
plt.show()
```
